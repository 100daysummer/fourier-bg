# Inverse Discrete Fourier Transform
Let's start with seeing if we can render each of the 'frequencies' that the DCT gives us.

```python
freq = np.zeros((8, 8))
freq[0,0] = 1
freq
```

```python
freq_img = idct(idct(freq, axis=0), axis=1)
plt.imshow(freq_img, cmap='gray')
```

```python
# Ok, lets try for a more interesting frequency
freq = np.zeros((8, 8))
freq[2,3] = 1
freq_img = idct(idct(freq, axis=0), axis=1)
plt.imshow(freq_img, cmap='gray')  # Side point: this image looks wiggly haha
```
